```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

df = pd.read_csv('/content/Students Performance in Exams.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df['gender'].value_counts()
```

```python
plt.figure(figsize= (7,7))
plt.title('Gender v/s average score')
sns.barplot(x = 'gender', 
            y = 'average score',
            data = df);
```

Race/Ethnicity

```python
df['race/ethnicity'].value_counts()
```

```python
plt.figure(figsize= (7,7))

plt.title('Race/ethnicity v/s average score')
sns.barplot(x = 'race/ethnicity', 
            y = 'average score', 
            hue = 'gender',
            palette = 'hls',
            data = df);
```

```python
df['parental level of education'].value_counts()
```

```python
plt.figure(figsize= (7,7))
plt.title('Parental level of education v/s average score')
plt.xticks(rotation = 90)
sns.barplot(x = 'parental level of education', 
            y = 'average score', 
            data = df,
            hue = 'gender',
            palette = 'tab10');
```

```python
df['test preparation course'].value_counts()
```

```python
plt.figure(figsize= (7,7))

plt.title('test preparation course v/s v/s average score')

sns.barplot(x = 'test preparation course', 
            y = 'average score', 
            data = df);
```
